2. Import the q2data.csv dataset.
Tidy data before importing dataset because k is a non-numerical indicator and converting it to a numerical indicator.

```python
q2data = pd.read_csv('Data/q2data.csv')
q2data = q2data.dropna()

def convert_k_to_numeric(value):
    if isinstance(value, str) and 'k' in value:
        return float(value.replace('k', '')) * 1000
    return value 

q2data['energy'] = q2data['energy'].apply(convert_k_to_numeric).astype(int)

q2data.head()
q2data
```

3. Recreate the plot as best you can using plotnine.

Sources:
https://plotnine.org/reference/guide.html

https://ggplot2.tidyverse.org/reference/ggtheme.html

```python
from plotnine import (
    ggplot, aes, geom_point, scale_x_continuous, scale_y_continuous,
    theme, element_rect, element_line, scale_size_continuous, 
    scale_fill_manual, labs, guides, annotate
)

(
    ggplot(q2data, aes(x='exports', y='imports', fill='four_regions', size='energy'))
    + geom_point()
    + scale_x_continuous(
        breaks=[20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220],
        limits=[0, 220]
    )
    + scale_y_continuous(
        breaks=[50, 100, 150, 200, 250, 300, 350, 400],
        limits=[0, 400]
    )
    + theme(
        panel_background=element_rect(fill='white'),
        figure_size=(8, 4),
        axis_line_x=element_line(color="gray"),
        axis_line_y=element_line(color="gray"),
        panel_grid_major=element_line(color="#D3D3D3")
    )
    + scale_size_continuous(range=(0.01, 8))  # Changed to continuous
    + scale_fill_manual(values=['#00FFFF', '#52D017', '#F67280', '#FFEF00'])
    + labs(x='Exports (% of GDP)', y='Imports (% of GDP)', fill='Four Regions')
    + guides(size=False)
    + annotate(
        'text', x=110, y=175, label='1997', size=150, 
        color='gray', alpha=0.2, family='Arial'
    )
)
```

This plot sacrifices some of the nuance in the data compared to scatterplots, which allow for more precise exploration of individual data points.
