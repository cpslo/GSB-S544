##**Question 1.2 [10 points]**

Many National Parks have interesting names. Which ones have names starting with the rare letters: V, W, X, Y, or Z?

```python
# Step 1: First, I would define a function to apply boolean masking, which searches for park names that start with the letters (V, W, X, Y, or Z).
# The function uses .startswith() combined with a tuple of letters for this purpose.
rare_letters = ['V', 'W', 'X', 'Y', 'Z']
def starts_with_rare_letter(area_name):
    return area_name.startswith(tuple(rare_letters))
```

By using `.startswith()` and `tuple()`, we determine:

- **`.startswith()` ensures case matching**: It checks if the string starts with the specified letters exactly as they appear, making it efficient for direct comparisons.  
- **`tuple()` allows for multiple starting letters**: It enables checking against multiple prefixes (e.g., V, W, X, Y, Z) simultaneously, streamlining the process without requiring multiple conditions or loops.



```python
# Step 2: Next, I would apply the function to create a Boolean mask
mask = df['area_name'].apply(starts_with_rare_letter)
mask
```

By using `.apply()`and creating a Boolean Masking:

*   `.apply()` applies a function (`starts_with_letters`) to each element of the name column in the dataset

*   The function checks if the first character of the name matches any of the rare letters.

*  A Boolean mask is a series of True and False values, where True indicates rows that meet a specific condition.

*  In this case, the mask checks if each park name starts with V, W, X, Y, or Z.


```python
# Step 3: After, I would use the mask to filter and extract park names

rare_letter_parks = df[mask]  # Filters the dataframe using the mask
matching_parks = rare_letter_parks['area_name'].unique() # Extracts unique park names
```

By using `.unqiue()`, we can:


*   Retrieve all distinct (non-duplicate) values from the specified column, ensuring that each park name appears only once in the output.
*    If a park name appears multiple times in the dataset, .unique() prevents redundancy in the final result, providing a clean list of unique park names that meet the condition.



```python
# Step 4: Lastly, I would display the filtered results
print("National Park Names Starting with V, W, X, Y, or Z:")
print(matching_parks)
```

Colorado, Alaska, and California also pride themselves on having long trails. Suppose we define a "long trail" as one that is more than ten kilometers. Do these three states have an equal distribution of long trails and short trails?

```python
# Step 1: First, I would use Boolean filtering for states Colorado, Alaska, and California

three_states = df_parks[df_parks['state_name'].isin(['Colorado', 'Alaska', 'California'])]
```

```python
# Step 2: Next, I will convert length from meters to kilometers
# Received SettingwithCopyWarning used .loc to fix
three_states.loc[:,'length_km'] = three_states['length'] / 1000
```

```python
# Step 3: Next, I will classify for long and short traails
three_states.loc[:,'trail_type'] = np.where(three_states['length_km'] > 10, 'Long', 'Short')
```

```python
 # Step 4: Next, I will find the distribution of long and short trails by using pd.crosstab() to calculate the counts of each type for three states
 contingency_table = pd.crosstab(
    three_states['state_name'],
    three_states['trail_type']
)
```

```python
# Step 5: Lastly, I would display the total distribution of long vs. short trail results for the states
print("Trail Distribution (Long vs. Short) for Each State:")
contingency_table
```

```python
three_states['length_km']
```

```python
# Step 5: I would display the distribution in percentages to interpret the equal distribution
trail_distribution_percentage = contingency_table.div(contingency_table.sum(axis=1), axis=0) * 100


trail_distribution_percentage
```



### **Interpretation**
The distribution of long and short trails is **not equal** across the three states:
1. **Alaska** has the highest percentage of short trails (71.43%) and the lowest percentage of long trails (28.57%).

The significant variation in these percentages demonstrates that the three states do not have an equal distribution of long and short trails.

The plot below visually answers Quesion 1.4. Describe how you would recreate this plot making reference to elements of the *grammar of graphics*.

```python
from plotnine import ggplot, aes, geom_bar, labs, theme_minimal, theme, element_text, scale_y_continuous, scale_fill_manual

# Step 1: First, I would convert the contingency table to proportions.
# This would normalize the counts in the contingency table to create proportions in decimal form. This ensures the y-axis reflects proportions ranging from 0 to 1.

# Step 2: Then, I would reshape the data for plotting.
# I would use the .melt() function to reshape the contingency table into a format suitable for plotnine, creating columns for state_name, trail_type, and proportion.
trail_distribution_percentage = contingency_table.div(contingency_table.sum(axis=1), axis=0)
trail_distribution_melted = trail_distribution_percentage.reset_index().melt(id_vars='state_name', var_name='trail_type', value_name='proportion')

# Step 3: Then, I would use plotnine map aesthetics for the plot.
# I would define the x-axis (state_name), y-axis (proportion), and the fill color (trail_type) for differentiating between long and short trails.

plot = (
    ggplot(trail_distribution_melted, aes(x='state_name', y='proportion', fill='trail_type'))

# Step 4: I would add bar chart layer.
# I would use Use geom_bar() to create a stacked bar chart where the bars represent proportions of trail types.
    + geom_bar(stat='identity', position='stack')
# Step 5: I would add labels and titles.
    + labs(
        title="Long vs Short Trails in CA, CO, and AK",
        x="State",
        y="Proportion",
        fill="Trail Type"
    )
# Step 6: I would use scale_y_continuous() to set the y-axis ticks to proportions (0, 0.25, 0.50, 0.75, 1.0).
    + scale_y_continuous(breaks=[0, 0.25, 0.50, 0.75, 1.0])
# Step 7: I would assign custom colors for trail types.
    + scale_fill_manual(values=['#e85ace', '#4682B4'])
# Step 8: I would use Use theme_minimal() to create a clean design and customize elements such as rotated x-axis labels for readability.
    + theme_minimal()
    + theme(
        axis_text_x=element_text(angle=45, hjust=1),  # Rotate x-axis labels
        plot_title=element_text(size=16),
        axis_title=element_text(size=14),
        legend_title=element_text(size=12)
    )
)

# Display the plot
plot
```

#### **Question 2.2 [10 points]**

Describe what could happen when pd.merge() combines two datasets on a single variable, but
there are other common variables between the two datasets.

*Corrected Answer: Received Partial Credit*


When `pd.merge()` combines two datasets on a single key variable, it joins rows based on that shared variable (e.g., an ID). Variables that are joined on appear only once in the final DataFrame, but other common variables (not used for the join) will appear twice with suffixes `_x` and `_y` to differentiate them. For example, if both datasets have a column `Count`, the result will include `Count_x` from the left DataFrame and `Count_y` from the right. These suffixes can be customized using the `suffixes=` parameter to make the output clearer.

with the term *tidy data*?

*Corrected Answer: Received Partial Credit*


Long and wide data formats differ in how they are arranged by columns, rows, and cells. Long formats typically have fewer columns and more rows, while wide formats have more columns and fewer rows. Long data formats are more associated with *tidy data* because they follow the required rules where columns represent variables, rows represent observations, and cells contain single values. In contrast, wide data formats often need to be transformed into a long format (e.g., using `pd.melt()`) to align with the principles of tidy data for effective analysis or visualization.
