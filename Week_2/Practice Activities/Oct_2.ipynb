# Merge/Stacking

Adding rows below the 1st dataset

```python
names1995.merge(names2015, on=["Name", "Sex"])
```

```python
names1995.merge(names2015, on=["Name", "Sex"], suffixes=("1995", "2015"))
```

The above code merges the combination of Name and Sex and finds it in the 2015 and 1995 dataset. The "suffixes" code labels the Count_X and Count_Y to interpret the year.


The most important component of merging two datasets is the presence of at least one key variable that both datasets share.

5.3.2 Types of Joins


```python
data_dir = "http://dlsun.github.io/pods/data/names/"

names1995 = pd.read_csv(data_dir + "yob1995.txt",
                        header=None, names=["Name", "Sex", "Count"])
names2005 = pd.read_csv(data_dir + "yob2005.txt",
                        header=None, names=["Name", "Sex", "Count"])
```

```python
# outer join
names_outer = names1995.merge(names2005, on=["Name", "Sex"], how="outer")
names_outer
```

Since, outer join merges the two datasets not accounting for the missing values. 
What does Python do with missing values? 

```python
# left join
names_left = names1995.merge(names2005, on=["Name", "Sex"], how="left")
names_left
```
