```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_temps(month=None):
    base_url = "https://www.extremeweatherwatch.com/us-state-averages"

    # Construct URL based on the month provided
    if month is not None:
        my_url = base_url + "/month-" + month
    else:
        my_url = base_url

    # Fetch the page content
    response = requests.get(my_url)

    # Parse the page content
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the first table in the page
    table = soup.find_all("table")[0]

    rows = []

    # Iterate over the table rows, skipping the header
    for row in table.find_all("tr")[1:]:
        cells = row.find_all('td')

        # Extract relevant data from the cells
        state = cells[0].text.strip()
        high = cells[1].text.strip()
        low = cells[2].text.strip()
        prep = cells[3].text.strip()

        # Append a dictionary of the row data
        rows.append({
            'state': state,
            'high': high,
            'low': low,
            'precipitation': prep
        })

    return pd.DataFrame(rows)
```

```python
#2.2

import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_temps(month=None):
    base_url = "https://www.extremeweatherwatch.com/us-state-averages"

    # Construct URL based on the month provided
    if month is not None:
        my_url = base_url + "/month-" + month
    else:
        my_url = base_url

    # Fetch the page content
    response = requests.get(my_url)

    # Parse the page content
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the first table in the page
    table = soup.find_all("table")[0]

    rows = []

    # Iterate over the table rows, skipping the header
    for row in table.find_all("tr")[1:]:
        cells = row.find_all('td')

        # Extract relevant data from the cells
        state = cells[0].text.strip()
        high = cells[1].text.strip()
        low = cells[2].text.strip()
        prep = cells[3].text.strip()

        # Append a dictionary of the row data
        rows.append({
            'state': state,
            'high': high,
            'low': low,
            'precipitation': prep
        })

    return pd.DataFrame(rows)
```

```python
#2.3 - pd.merg

# Filter trails based on the given criteria
recommended_trails = recommend_trails(df_parks,
                                      min_elev_gain=1000,
                                      max_elev_gain=3000,
                                      min_rating=4.5,
                                      route_type='Loop')

# Display the recommended trails
print(recommended_trails)

# Get temperature data for May through August
temp_may = scrape_temps('may')
temp_june = scrape_temps('june')
temp_july = scrape_temps('july')
temp_august = scrape_temps('august')

# Merge monthly temperatures into one dataframe
temp_data = pd.concat([temp_may, temp_june, temp_july, temp_august])

# Merge trail data with temperature data on state
merged_data = pd.merge(recommended_trails, temp_data, left_on='state_name', right_on='state')

# Display the merged dataset
print(merged_data)

# Convert elevation gain to feet
merged_data['elevation_ft'] = merged_data['elevation_gain'] * 3.28084

# Adjust temperatures for elevation using lapse rate (5.4°F per 1000 feet)
merged_data['adjusted_high'] = merged_data['high'].astype(float) - (merged_data['elevation_ft'] / 1000) * 5.4
merged_data['adjusted_low'] = merged_data['low'].astype(float) - (merged_data['elevation_ft'] / 1000) * 5.4

# Display the final dataset with adjusted temperatures
print(merged_data[['trail_id', 'state', 'elevation_gain', 'avg_rating', 'adjusted_high', 'adjusted_low']])
```
